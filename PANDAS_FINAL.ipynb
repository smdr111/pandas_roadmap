 **34**. Endi ushbu DataFrame uchun `adjacent` nomli yangi ustun yarating. Ushbu ustunda tarmoqdagi qo驶shni kvadratlarda topilgan mine始lar soni bo驶lishi kerak.

(Masalan, `(0, 0)` koordinatasi uchun yozuv bo驶lgan birinchi qator uchun `(0, 1)`, `(1, 0)` va `(1) koordinatalarida qancha mine borligini hisoblang. , 1)`.)

````python
nbr = df.merge(df, how='cross', suffixes=('', '_n'))
````

````python
mask = (
    (nbr['X'].sub(nbr['X_n']).abs() <= 1) &
    (nbr['Y'].sub(nbr['Y_n']).abs() <= 1) &
    ~((nbr['X'] == nbr['X_n']) & (nbr['Y'] == nbr['Y_n']))
)
````

````python
adj = (nbr.loc[mask]
          .groupby(['X','Y'])['mine_n']
          .sum()
          .reset_index()
          .rename(columns={'mine_n': 'adjacent'}))
````

````python
df = df.merge(adj, on=['X','Y'], how='left')
````

````python
df
````
